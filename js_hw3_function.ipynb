```python
import datetime

wns = ['Mon.', 'Tue.', 'Wed.', 'Thu.', 'Fri.', 'Sat.', 'Sun.']
#weekday Monday is 0
def printMonth(year, month):
    a = datetime.datetime(year,month,1)
    print(f"{a.year} 年 {a.month} 月")

    if a.year == 12:
        b = datetime.datetime(a.year+1, 1, 1)
    else:
        b = datetime.datetime(a.year, a.month+1, 1)

    wa = a.weekday()
    wb = b.weekday()

    for i in wns:
        print(f"{i:}", end=" ")
    print("")
    #for i in range(wa):
    print(f"{" ":<5}"*(wa), end="")

    d = 0
    while True:
        d += 1
        if wa == 7:
            wa = 0
            print("")
        if d > 28 and wb == wa:
            break

        wa += 1
        print(f"{d:<5}", end="")

        if d == 40:
            break

printMonth(2026,4)
```
