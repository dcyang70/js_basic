```python
month = int(input("please enter a month:"))
#print(month//3)
if month > 0 and month < 13:
    match month//3:
        case 1:print("spring")
        case 2:print("summer")
        case 3:print("fall")
        case _:print("winter")
```

```python
hours = int(input("working hours:"))
print(hours)
PAY60 = 120 #7200
PAY80 = 1.25 * PAY60 #  150 3000 
PAY_OVER_80 = 1.5 * PAY60 #180
total = 0
if hours > 80:
    total = (hours-80) * PAY_OVER_80 + 20*PAY80 + 60*PAY60
    pass
elif hours > 60 :
    total = (hours-60) * PAY80 + 60*PAY60
    pass
else:
    total = hours * PAY60
    pass

print(f"The pay is {total}")
```

```python
amount = int(input("the amount:"))
money = int(input("you pay:"))
print(f"amount:{amount} money:{money}")
bills = [1000, 500, 100, 50, 10, 5] #1
num = 0
if amount > money:
    print("money is not enough")
elif amount == money:
    print("No need for change")
else:
    money = money-amount
    msg = ""
    for i in bills:
        if money > i:
            num = money // i
            money = money % i
            msg = f"{msg} ${i}x{num},"
    if money > 0:
        msg = f"{msg} $1x{num},"
    print(msg)
        
```
